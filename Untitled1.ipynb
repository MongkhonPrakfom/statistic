```python
import numpy as np
import matplotlib.pyplot as plt #untuk generate graph
import pandas as pd #untuk table or data frame

#left-skewed min kecil pada median
#right-skewed min besar pada median

data = np.array([56, 67, 45, 70, 89, 90, 78, 65, 72, 60, 55, 80, 85, 62, 74])
```

```python
sort = np.sort(data)
print("Sorted :", sort)
```

```python

mean = np.mean(data) #powerfull tetapi sensitive
print("Mean :", mean)
```

```python
median = np.median(data)
print("Median :", median)
```

```python
variance = np.var(data)
print("Variance :", variance)
```

```python
plt.hist(data, bins = "auto")
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```
